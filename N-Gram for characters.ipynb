```python
import random
```

```python
text = """Global warming or climate change has become a worldwide concern. It is gradually developing into an unprecedented environmental crisis evident in melting glaciers, changing weather patterns, rising sea levels, floods, cyclones and droughts. Global warming implies an increase in the average temperature of the Earth due to entrapment of greenhouse gases in the earthâ€™s atmosphere."""
```

```python
# building the model 
n = 5
ngrams = {} # defining the empty dictionary 
for i in range(len(text) - n):
    gram = text[i:i+n] # 0:3 'glo'
    print(gram)
    if gram not in ngrams.keys():
        ngrams[gram] = [] # initializing the empty string 
    ngrams[gram].append(text[i+n])# 0+3 = 3 in global index 3 is 'b'
    print(ngrams)
    
    
    
    
```

```python
# testing the model 
currentgram = text[0:n] # 'glo'
result = currentgram
for i in range(100):# we want a sentence of length 100 characters including space
    if currentgram not in ngrams.keys():# if glo is not in text
        break #moves out of the loop
    possibilities = ngrams[currentgram] # what are the ngrams of 'glo'  -- ['b']
#     print(possibilities)
    nextitem = possibilities[random.randrange(len(possibilities))]# this will randomly pick if the list is more than 1
#     print(nextitem)
    result  += nextitem # add the next to the result, so the result is glob
    currentgram = result[len(result)-n: len(result)]# update the currentgram 
    #len(glob) = 4; currentgram = result[1:4] = 'lob'
    print(currentgram)

    
    
```
